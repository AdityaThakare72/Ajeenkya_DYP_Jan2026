```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([12, 23, 34, 37, 40, 44, 52, 57, 68, 72])
y = np.array([15, 22, 33, 35, 45, 43, 50, 60, 65, 70])
z = np.array([71, 66, 59, 54, 48, 42, 38, 33, 25, 20])
```

# line chart

```python
plt.xlabel('x values')
plt.ylabel('y values')
plt.title("line chart")
plt.plot(x, y, marker = 'o', mfc = 'orange', ls = '--');
plt.grid();
```

```python
plt.xlabel('x values')
plt.ylabel('y values')
plt.title("line chart")
plt.plot(x, z, marker = 'o', mfc = 'orange', ls = '--');
plt.grid();
```

```python
plt.xlabel('x values')
plt.ylabel('y values')
plt.title("line chart")
plt.plot(x, y, marker = 'o', mfc = 'orange', ls = '--',
         label = "rice");
plt.plot(x, z, marker = 'o', mfc ="" 'blue', ls = '--',
         color = 'black', label = "sugar");
plt.grid();
plt.legend();
```
